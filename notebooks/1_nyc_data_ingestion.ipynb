```python
def make_download_url(dataset_id):
    return f"https://data.cityofnewyork.us/api/views/{dataset_id}/rows.csv?accessType=DOWNLOAD"

rows = []

# Education
for ds_id in education_ids:
    rows.append({
        "dataset_id": ds_id,
        "category": "Education",
        "name": None,  # optional, can fill later
        "nyc_url": f"https://data.cityofnewyork.us/resource/{ds_id}",
        "csv_url": make_download_url(ds_id),
        "local_path": f"data/raw/education/{ds_id}.csv"
    })

# City Government
for ds_id in city_gov_ids:
    rows.append({
        "dataset_id": ds_id,
        "category": "City Government",
        "name": None,
        "nyc_url": f"https://data.cityofnewyork.us/resource/{ds_id}",
        "csv_url": make_download_url(ds_id),
        "local_path": f"data/raw/city_gov/{ds_id}.csv"
    })

dataset_index = pd.DataFrame(rows)
dataset_index.head()
```

```python
import requests
import time


def download_csv(url, local_path, timeout=10):
    """
    Attempts to download a CSV with a hard timeout.
    If it takes longer than `timeout` seconds, it fails and moves on.
    """
    try:
        resp = requests.get(url, timeout=timeout, stream=True)
        resp.raise_for_status()

        # Write file in chunks (safer for large files)
        with open(local_path, "wb") as f:
            for chunk in resp.iter_content(chunk_size=8192):
                f.write(chunk)

        return "ok"

    except requests.exceptions.Timeout:
        print(f"Timeout after {timeout} seconds: {url}")
        return "timeout"

    except Exception as e:
        print(f"Failed to download {url}: {e}")
        return "error"



statuses = []

for i, row in dataset_index.iterrows():
    print(f"[{i+1}/{len(dataset_index)}] Downloading {row['dataset_id']} ...")
    status = download_csv(row["csv_url"], row["local_path"], timeout=10)
    statuses.append(status)

dataset_index["download_status"] = statuses
dataset_index.head()
```

```python
can_load = []
num_cols_sample = []

for i, row in dataset_index.iterrows():
    try:
        df = pd.read_csv(row["local_path"], nrows=100)
        can_load.append(True)
        num_cols_sample.append(df.shape[1])
    except Exception as e:
        print(f"Load failed for {row['dataset_id']}: {e}")
        can_load.append(False)
        num_cols_sample.append(None)

dataset_index["can_load_100_rows"] = can_load
dataset_index["num_cols_sample"] = num_cols_sample

dataset_index.head()


```
